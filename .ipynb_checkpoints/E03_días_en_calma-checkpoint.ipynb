```python
# importamos las librerías que vamos a necesitar
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import numpy as np


# nuestra librería
import ist_lib as ist

folder='dst_stations/2015-03/'
month='201503'

# observatorios usados para el índice Dst
obs=['her','hon','kak','sjg']
# días a observar
qs=['10','30','05','14','09']

# variable para guardar todos los datos (1440 minutos) de los cuatro observatorios,
# los cinco días en calma
dat = np.zeros([len(obs),len(qs),1440])

# leer todos los días en calma del mes para cada observatorio (sólo la componente H (c1))
for o in obs:
    for q in qs:
        filename=folder + o + '/' + o + month + q + 'qmin.min'
        df=ist.read_im(filename=filename)
        data=df['data']
        dat[obs.index(o)][qs.index(q)][:]=data.c1
```

```python
# primer observatorio HER, tercer día 05 (05-03-2015)
plt.plot(dat[0][2][:]);
```

## P5: ¿Podrías indicar una forma para determinar en qué minuto empieza el día local de cada obseratorio?
### R5:

```python
# Obtenemos las longitudes de los cuatro observatorios
lons=['','','','']
for o in obs:
        filename=folder + o + '/' + o + month + q + 'qmin.min'
        df=ist.read_im(filename=filename)
        lons[obs.index(o)]=df['lon']
print(obs)
print(lons)
```

### Si observamos las longitudes tienen valores mayores de 180, es decir, entre 0 y 360. 
### Convirtamos las longitudes mayores de 180 grados en negativas, es decir, entre -180 y 180:

```python
lons = [(float(lon) + 180) % 360 - 180 for lon in lons]
print(obs)
print(lons)
```

### Vamos a rotar los datos un número de minutos correspondiente a su posición en longitud

```python
# Obtenemos un día de HON y lo mostramos en la gráfica
# estos datos están centrados en el día con tiempo universal UT
hon=dat[1][2][:]
plt.plot(hon, label='UT');

# rotamos los datos el número de minutos correspondientes a su longitud
lon=lons[1]
mins=int(lon / 360 * 1440)
hon_lt=np.roll(hon,mins)

# mostramos los datos rotados, es decir, centrado en el día local
plt.plot(hon_lt, label='LT');
print(mins)

plt.legend()
```
