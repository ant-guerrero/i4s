```python
# importamos las librerías que vamos a necesitar
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import numpy as np

# nuestra librería
import ist_lib as ist

folder='dst_stations/2015-03/'
month='201503'

lat=0.
lon=0.
comp='---'

# observatorios usados para el índice Dst
obs=['her','hon','kak','sjg']

# días a observar: todo el mes '01' - '31'
qs=pd.Series(range(1,32))
qs=pd.Series.tolist(qs.apply(lambda x: '{0:0>2}'.format(x)))
print(qs)

# variable para guardar todos los datos (1440 minutos) de los cuatro observatorios,
# los días necesarios
dat = np.zeros([len(obs),len(qs),1440])

# leer todos los días en calma del mes para cada observatorio (sólo la componente H (c1))
for o in obs:
    for q in qs:
        filename=folder + o + '/' + o + month + q + 'qmin.min'
        df=ist.read_im(filename=filename, lat=lat, lon=lon, comp=comp)
        dat[obs.index(o)][qs.index(q)][:]=df.c1
```

# Analicemos los datos de días consecutivos

```python
# Todos los días de HER
plt.plot(dat[0][:][:].reshape(-1));
```

```python
# Todos los días de HON
plt.plot(dat[1][:][:].reshape(-1));
```

```python
# Todos los días de KAK
plt.plot(dat[2][:][:].reshape(-1));
```

Para SJG vemos que sucede algo que no ha ocurrido antes:

```python
# Todos los días de SJG
plt.plot(dat[3][:][:].reshape(-1));
```

Este comportamiento se debe a que existen valores que se marcan como no válidos con el valor 99999.
Para poder representarlo bien debemos ignorar dichos valores. Uno de las formas de hacerlo es cambiar el valor 99999 por el indicador "not a number, NaN" (np.nan) como se muestra a continuación.

```python
# Todos los días de SJG
d=dat[3][:][:].reshape(-1)
d=[np.nan if x==99999 else x for x in d]
plt.plot(d);
```
