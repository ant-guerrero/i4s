# L01: Read files from intermagnet

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
```

```python
# we will work with the following file:
    # observatory: hermanus (HER)
    # day: 10-03-2013
    # type: definitive (d)
    # resolution: minutes (min)
filename='i4s_geomag_data/her20130310dmin.min'
```

```python
# read the file
names=['diahora','doy','c1','c2','c3','c4']
colspecs=[(0,23), (25,27), (32,40), (42,50), (52,60), (62,70)]
df=pd.read_fwf(filename,header=26,names=names, colspecs=colspecs)
```

```python
# crete the variables we will later use
tiempo=pd.to_datetime(df.diahora.apply(str), format = '%Y-%m-%d %H:%M:%S.%f')
dat1=df.c1   # component 1
dat2=df.c2   # component 2
dat3=df.c3   # component 3


# check size of variables
print(dat1.size)
print(dat2.size)
print(dat3.size)
print(tiempo.size)
```

## Q04: ¿What components are we observing?
